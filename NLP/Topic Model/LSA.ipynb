```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import string
import gensim.corpora as corpora
import gensim
from gensim.models import LsiModel
```

```python
nltk.download('punkt')
nltk.download('wordnet')   
nltk.download('stopwords')
```

```python
data = pd.read_csv("news_articles.csv")
data.head()
```

```python
articles = data["content"]
```

```python
def text_pre_process(s: str):

    #lower
    s = s.lower()

    #remove punctuation
    s = s.translate(str.maketrans('', '', string.punctuation))
    
    #tokenization
    tokens = word_tokenize(s)

    #remove stop word
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    #lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    return tokens
    
    
    
```

```python
articles = articles.apply(text_pre_process)
```

```python
#dictionary
id2word = corpora.Dictionary(articles)
id2word
```

```python
corpus = [id2word.doc2bow(text) for text in articles]
```

```python
lsa_model = LsiModel(corpus= corpus, id2word= id2word, num_topics= 3)
```

```python
topics = lsa_model.print_topics(num_topics= 3, num_words= 5)
```

```python
for topic in topics:
    print(topic)
```
