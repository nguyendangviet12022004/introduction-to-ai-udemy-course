```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import string
import gensim.corpora as corpora
import gensim
from gensim.models import LsiModel
from gensim.models.coherencemodel import CoherenceModel
import matplotlib.pyplot as plt
```

```python
nltk.download('punkt')
nltk.download('wordnet')   
nltk.download('stopwords')
```

```python
data = pd.read_csv("news_articles.csv")
data.head()
```

```python
articles = data["content"]
```

```python
def text_pre_process(s: str):

    #lower
    s = s.lower()

    #remove punctuation
    s = s.translate(str.maketrans('', '', string.punctuation))
    
    #tokenization
    tokens = word_tokenize(s)

    #remove stop word
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    #lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    return tokens
    
    
    
```

```python
articles = articles.apply(text_pre_process)
```

```python
#dictionary
id2word = corpora.Dictionary(articles)
id2word
```

```python
corpus = [id2word.doc2bow(text) for text in articles]
```

```python
min_num_topics = 3
```

```python
max_num_topics = 12
```

```python
lsa_models = []
coherence_scores = []
```

```python
for i in range (min_num_topics, max_num_topics + 1):
    lsa_model = LsiModel(num_topics= i, id2word=id2word, corpus=corpus)
    lsa_models.append(lsa_model)
    coherence_scores.append(CoherenceModel(model=lsa_model, dictionary = id2word , coherence='c_v', texts=articles).get_coherence())
```

```python
xpoints = [x for x in range(min_num_topics, max_num_topics + 1)]
ypoints = coherence_scores
plt.plot(xpoints, ypoints)
plt.show()
```
