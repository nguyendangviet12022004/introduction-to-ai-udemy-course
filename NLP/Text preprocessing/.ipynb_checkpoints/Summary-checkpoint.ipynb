```python
import pandas as pd
import nltk 
from nltk.corpus import stopwords
import re
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import numpy as np
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df = pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
df["review_lowercase"] = df.apply(lambda x: x["Review"].lower(), axis= 1)
```

```python
def remove_stop_words(s: str):
    return ' '.join([c for c in s.split() if c not in stop_words])
```

```python
df["review_no_stopwords"] = df.apply(lambda x : remove_stop_words(x["review_lowercase"]) , axis=1)
```

```python
def remove_punctuation(s: str):
    return re.sub(r"[*]", "star", s)
```

```python
df["review_no_stopwords_no_punct"] = df.apply(lambda x : remove_punctuation(x["review_no_stopwords"]) , axis=1)
```

```python
df["tokenized"] = df.apply(lambda x : word_tokenize(x["review_no_stopwords_no_punct"]), axis= 1)
```

```python
pd = PorterStemmer()
```

```python
df["stemmed"] = df.apply(lambda x : [pd.stem(c) for c in x["tokenized"] ], axis= 1)
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
df["lemmatized"] = df.apply(lambda x :  [lemmatizer.lemmatize(c) for c in x["tokenized"] ], axis= 1)
```

```python
clean_tokens = [token for block in df['lemmatized'] for token in block]
```

```python
unigrams = pd.Series(nltk.ngrams(clean_tokens, 1)).value_counts()
```

```python
unigrams
```

```python
bigrams = pd.Series(nltk.ngrams(clean_tokens, 2)).value_counts()
```
