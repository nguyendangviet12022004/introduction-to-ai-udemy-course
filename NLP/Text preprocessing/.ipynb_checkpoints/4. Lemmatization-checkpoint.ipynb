```python
import nltk
```

```python
nltk.download("wordnet")
```

```python
from nltk import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
test_words = [
    "run", "running", "ran",             
    "leaf", "leaves",                    
    "good", "better", "best",           
    "am", "is", "are", "was", "were",    
    "feet", "foot",                      
    "studied", "studies", "studying",    
    "mice", "mouse",                     
    "harder", "hardest"                  
]
```

```python
result_words = [lemmatizer.lemmatize(word) for word in test_words ]
```
