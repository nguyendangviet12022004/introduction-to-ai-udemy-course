```python
list(nltk.ngrams(tokens,2))
```

```python
bigrams = pd.Series(nltk.ngrams(tokens, 2)).value_counts()
```

```python
bigrams
```

```python
bigrams[:10].sort_values().plot.barh(
    title="Most Frequently Occuring Unigrams",
    color='lightsalmon', 
    width=.9, 
    figsize=(12, 8),
    xlabel="Freq",
    ylabel="Word"
)
```
